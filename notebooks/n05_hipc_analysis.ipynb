```python
speedups_fp32 = []
speedups_fp64 = []
def amp_fp32_diff(ds):
#     print(ds)
#     print(ds.loc[ds.dtype=="a", "runtime"].iloc[0])
#     print(ds.loc[ds.dtype=="fp32", "runtime"].iloc[0])
#     speedups_fp32.append(100*(fp32_time-amp_time)/fp32_time)
#     speedups_fp64.append(100*(fp64_time-amp_time)/fp64_time)
    
    try:
        # fp32
        amp_time = ds.loc[ds.dtype=="a", "runtime"].iloc[0]
        fp32_time = ds.loc[ds.dtype=="fp32", "runtime"].iloc[0]
        
        # fp64
        fp64_time = ds.loc[ds.dtype=="fp64", "runtime"].iloc[0]
        
        # append to list
        speedups_fp32.append(fp32_time/amp_time)
        speedups_fp64.append(fp64_time/amp_time)
    except:
        print(ds.model.iloc[0])
#     return None
    
df.loc[df["batch_size"]==2048].groupby("model").apply(amp_fp32_diff)
```
